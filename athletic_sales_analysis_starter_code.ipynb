```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
as2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
as2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
as2020
```

```python
# Display the 2021 sales DataFrame
as2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
as2020.dtypes
```

```python
# Check the 2021 sales data types.
as2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
ascombined = pd.concat([as2020,as2021],axis=0,join='inner',ignore_index=True)
#ascombined.reset_index(inplace=True)  
ascombined       
```

```python
# Check if any values are null.
# first method: use count to see whether any columns have fewer values, which would indicate NaN values in the column
ascombined.count()
# by method 1, we can see that all columns have same number of values, meaning that there are no NaN values in the original data
```

```python
# second method: create a copy df where NaN rows are dropped.  compare the counts of this to the original
testdf = ascombined.dropna(how='any')
print('rows in ascombined',len(ascombined),', and rows that don\'t contain NaNs',len(testdf))
# by method 2, we can see that the original df has same number of rows as the df with NaNs removed, which means there are no NaNs in the original df
```

```python
# Check the data type of each column
print(f'The combined df has {len(ascombined)} rows. The original dfs have lengths {len(as2020)} and {len(as2021)} rows.')
ascombined.dtypes
# this shows us that the data types are same as before,  
```

```python
# Convert the "invoice_date" to a datetime datatype
# use datetime.strptime()
print('data type for invoice date prior to conversion is',ascombined['invoice_date'].dtypes)

# use the pandas function to convert the time field from string(object) to datetime
ascombined['invoice_date'] = pd.to_datetime(ascombined['invoice_date'])
ascombined
```

```python
# Confirm that the "invoice_date" data type has been changed.
print('data type for invoice date after conversion is',ascombined['invoice_date'].dtypes)
```
